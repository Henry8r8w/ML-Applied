## Random Sampling + Infinite Outcomes

Say, you toss a count **`twice`**, you have a sample space of
$$ \Omega = \{\text{HH}, \text{TT}, \text{HT}, \text{TH}\}$$

Now say you toss an infinite sqeuece of coin tosses

$$\Omega = \text{HTHHTHTHHTHTHT...}$$
- the sample space now becomes a **`uncountable`** set


## Inclusion and Exclusion Principle
### Probability Perspective
Let X be the number of the first toss that is **`H`** and the rest being **`T`**

Say, we are interested in inducing the probabilitiy  where event X has **`1 Head`** and **`2+ Tails`**

$$P(X = 3) = P(TTH) = \frac{1}{8}$$

$$P(X \geq 3) = P(\text{TTH, TTTH, TTTTH,...}) = 1/8 + 1/ 16 + 1/ 32 ... = 1/4$$

- $\{X \geq 3\} = \{X < 3\}^c$
- $P(X < 3) = P(X \leq 2) = P(H, TH) = P(\text{H}) + P(\text{TH}) = \frac{1}{2} + \frac{1}{4}$
- $P( X \geq 3) = 1 - P( X < 3 ) = 1 - \frac{3}{4} = \frac{1}{4}$

## Example
Sample an integer from the set $\{1,2,3,..., 100\}$. What is the probability that it is not divisible by 3 and not divisible by 3?


$$A = \{\text{num } \% \ 2 == 0\}$$
$$B = \{\text{num } \% \ 3 == 0\}$$

$$P(A^c \cap B^c) = P((A\cup B)^c)
= 1 - P(A \cup B) = 1 - (P(A) + P(B) - P(A \cap B))$$
- **Note**: the reason why we use union complement and rewrite the exrpession to get all the `True/non-C` conditions is becuase it is easier to compute
- $P(A) = (\{\text{largest 2n} / 2 \}) / (\text{sample space: 100})  = 100/ 2 / 100$
- $P(B) = (\{\text{largest 3n} / 3 \}) / (\text{sample space: 100}) = 99 / 3 / 100$
$ P(A \cap B) = (100/ 2 / 100) * (99/ 3 / 100) = 16.5/100 \equiv (\{\text{largest 6n} / 6 \}) / (\text{sample space: 100}) = 96/6/100$
- **Note**: you can also think of this probability of condition where the integer is divible by 6 (largest 6n /6/ 100)
  - P(A) * P(B) = P(A) AND P(B) only applied if A and B are mutually indepedent

$$P(A^c \cap B^c) = 1 - (\frac{1}{2} + \frac{33}{100} - \frac{16}{100}) = \frac{33}{100}$$

## DeMorgan Laws
**Two events**:
$$(A \cup B)^c = A^c \cap B^c$$

$$(A \cap B)^c = A^c \cup B^c$$

**Three events**:
$$(A \cup B \cup C)^c = A^c \cap B^c \cap C^c \cap D^c$$

$$(A \cap B \cap C)^c = A^c \cup B^c \cup C^c \cup D^c$$
