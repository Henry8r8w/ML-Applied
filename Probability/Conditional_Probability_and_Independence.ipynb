## Conditional Probability
**Definition**: If $P(B) > 0$ then the conditional probability of A igven B is

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

$$P(A) = \frac{\text{Area of } A}{\text{Area of } \Omega}$$

$$P(A|B) = \frac{\text{Area of } A\cap B}{\text{Area of }B}$$

## Independence
**Definition**: We say that A and B are independent if $P(A) = P(A|B)$

$$P(A) = P(A|B) = \frac{P(A \cap B)}{P(B)}$$

$$P(A)P(B) = P(A \cap B)$$
### Example: Symmetry of Independence
Toss a coin twice. By symmetry, HH, HT, TH, and TT are equally likely. Let $X_k$ denote k-th toss

$$P(X_2 = T| X_1 = H) = \frac{P(X_2 = T \text{ and } X_1 = H)}{P(X_1 = H)}$$
$$= \frac{P(HT)}{P(HH \text{ or } HT)} = \frac{1/4}{1/4 + 1/4} = 1/2$$

$$P(X_2 = T) = P(TT \text{ or } HT) = 2/4 = 1/2$$

We see that $P(X_2 = T| X_1 = H) = P(X_2 = T)$
- this makes  \{X_2 = T\} and \{X_1 = H\} indepedent

### Example

A test is 95% effective if the disease is present. It generate a false positive with probability of 1%. We will assume that 0.5% of the population has the disease. If the test is positive, what is the probability that in fact the person has the disease


Let A = \{the test is positive \} and D = \{the person has the disease \}

$$P(D) = 0.005$$

$$P(A|D^c) = 0.01$$

$$P(A|D) = 0.95 $$

We want $P(D|A)$, which is obtainable by baye's theorem


$$
P(D|A) = \frac{P(A|D)P(D)}{P(A|D)P(D) + P(A|D^c)P(D^c)} = \frac{0.95 \cdot 0.005}{0.95\cdot 0.005 + 0.01 \cdot 0.995}$$

## Indepdenece: with exclusion/inclusion

If A and B are independent than $A^c$ and B are independent

$$P(A \cap B) = P(A) P(B)$$

$$B = (A \cap B) \cup (A^c \cap B), (A \cap B) \cap (A^c \cap B)$$


$$P(B) = P(A \cap B) + P(A^c \cap B)$$

$$P(A^c \cap B) = P(B) = P(A \cap B) = P(B) - P(A) P(B) = P(B)(1- P(A)) = P(B)P(A^c)$$

## Jointly Independent
Definition: we say that A_1, ..., A_n are jointly indepednet if for any subfamily of the events

$$
P(A_{k1} \cap A_{k2} \cap ... \cap A_{kj}) = P(A_{k1})P(A_{k2})...P(A_{kj})
$$

If the only thing that we know is that $P(A_jA_k) = P(A_j)P(A_k)$ for all $j \neq k$, then we call $A_1, A_2, ..., A_n$ pairwise independent

### Example
Consider a regular tetrahedron with color of red, blue and green on each side, and we roll

$$P(R) = 1/2 = P(B) = P(G)$$

$$P(R \cap B) = 1/4$$

$$P(R \cap B \cap G) = 1/4 \\ \text{ (pair-wise independent)}$$


$$P(R)P(B)P(G) = 1/8 \neq 1/4 = P(R \cap B \cap G)$$
